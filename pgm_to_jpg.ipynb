```python
import cv2
import os

# Define input and output folders
pgm_folder = '/content/drive/MyDrive/miniMIAS_Dataset'  # Folder where .pgm images are
jpg_folder = os.path.join('/content/drive/MyDrive/MIAS', 'Images')  # Folder to store .jpg images
os.makedirs(jpg_folder, exist_ok=True)

# Loop through all files in the dataset folder
for filename in os.listdir(pgm_folder):
    if filename.endswith('.pgm'):
        pgm_path = os.path.join(pgm_folder, filename)
        jpg_path = os.path.join(jpg_folder, filename.replace('.pgm', '.jpg'))

        # Read PGM image
        img = cv2.imread(pgm_path, cv2.IMREAD_GRAYSCALE)

        # Save as JPG with maximum quality
        cv2.imwrite(jpg_path, img, [int(cv2.IMWRITE_JPEG_QUALITY), 100])

print("Conversion complete. JPG images saved in:", jpg_folder)
```
