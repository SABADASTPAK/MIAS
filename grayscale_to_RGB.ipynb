```python
import cv2
import os
import glob

# --- Paths ---
input_folder = "/content/drive/MyDrive/MIAS/CLAHE"      # folder with grayscale images
output_folder = "/content/drive/MyDrive/MIAS/CLAHE_RGB"               # new folder for RGB images

os.makedirs(output_folder, exist_ok=True)

# --- Process all images ---
for img_path in glob.glob(os.path.join(input_folder, "*")):
    # Read grayscale image
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        continue

    # Convert to RGB by replicating channels
    rgb_img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)

    # Save with same filename in new folder
    filename = os.path.basename(img_path)
    save_path = os.path.join(output_folder, filename)
    cv2.imwrite(save_path, rgb_img)

print("✅ All grayscale images converted to RGB and saved in:", output_folder)
```
