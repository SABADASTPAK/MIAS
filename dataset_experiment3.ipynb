```python
import pandas as pd
import os
import shutil
import random

# Paths (change these to match your structure)
csv_path = '/content/drive/MyDrive/miniMIAS_Dataset/mias_info_with_yolo.csv'
images_folder = '/content/drive/MyDrive/MIAS/CLAHE_RGB'
labels_folder = '/content/drive/MyDrive/MIAS/Labels'
output_root = '/content/drive/MyDrive/yolo_dataset_clahe'

# Read CSV
df = pd.read_csv(csv_path)

# Filter: keep rows where Abnormality Type is NORM and CIRC
filtered_df = df[df['Abnormality Type'].isin(['NORM', 'CIRC'])]


# ✅ Group by ID (each ID may have multiple annotations, so keep them grouped)
grouped = filtered_df.groupby('ID')

# Count groups by Abnormality Type
norm_groups = grouped.filter(lambda g: (g['Abnormality Type'] == 'NORM').all())
circ_groups = grouped.filter(lambda g: (g['Abnormality Type'] == 'CIRC').all())

num_norm = norm_groups['ID'].nunique()
num_circ = circ_groups['ID'].nunique()

# Among CIRC → count benign vs malignant (based on Severity)
circ_benign = circ_groups[circ_groups['Severity'] == 'B']['ID'].nunique()
circ_malignant = circ_groups[circ_groups['Severity'] == 'M']['ID'].nunique()

print("✅ Distribution of groups (by ID):")
print(f"  NORM groups       : {num_norm}")
print(f"  CIRC groups total : {num_circ}")
print(f"     - Benign       : {circ_benign}")
print(f"     - Malignant    : {circ_malignant}")
```
