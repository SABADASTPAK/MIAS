```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="/content/drive/MyDrive/yolo_dataset_experiment2/data.yaml",
    epochs=200,
    imgsz=640,
    batch=16,
    name="experiment2",

    # Geometric Augmentations
    degrees=10,
    translate=0.1,
    scale=0.5,
    shear=2,
    perspective=0.0005,
    flipud=0.5,
    fliplr=0.5,
    mosaic=1.0,
    mixup=0.2,
    copy_paste=0.3,



    project="/content/drive/MyDrive/YOLO_results"
)
```

```python
from ultralytics import YOLO

# Load the trained model correctly (best.pt)
model = YOLO("/content/drive/MyDrive/YOLO_results/experiment2/weights/best.pt")

# Run inference
results = model.predict(
    source="/content/drive/MyDrive/yolo_dataset_experiment2/test/images",
    save=True,
    imgsz=640,
    project="/content/drive/MyDrive/YOLO_results",
    name="MIAS_inference_experiment2"
)
```
