```python
from google.colab import drive
import os
import urllib.request
import zipfile

# 1. Mount Google Drive
drive.mount('/content/drive')

# 2. Define dataset directory path inside your Drive
dataset_dir = '/content/drive/MyDrive/miniMIAS_Dataset'
os.makedirs(dataset_dir, exist_ok=True)

# 3. Download the dataset
url = 'http://peipa.essex.ac.uk/pix/mias/all-mias.tar.gz'
download_path = os.path.join(dataset_dir, 'all-mias.tar.gz')

print("Downloading MIAS dataset...")
urllib.request.urlretrieve(url, download_path)
print("Download complete.")

# 4. Extract the tar.gz file
import tarfile

print("Extracting dataset...")
with tarfile.open(download_path, 'r:gz') as tar:
    tar.extractall(path=dataset_dir)
print("Extraction complete.")

# Optional: remove the compressed file to save space
os.remove(download_path)
print("Compressed file removed.")
```
